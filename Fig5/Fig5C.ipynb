```julia
using DataFrames, StatsPlots, Plots.Measures
```

```julia
default(fontfamily="Helvetica", xtickfontfamily="Helvetica Bold", xtickfontsize = 10, xguidefontsize = 12, lw = 0, bar_width=0.7, framestyle = :box, alpha = 0.7)
```

```julia
cp_sites = [285.0, 0.0]
cp_site_proportion = cp_sites ./= sum(cp_sites)

mt_sites = [362.0, 6.0]
mt_site_proportion = mt_sites ./= sum(mt_sites)

sites = DataFrame(organelle = String[], direction = String[], proportion = Float64[])
push!(sites, ("cp", "CtoU", cp_site_proportion[1]))
push!(sites, ("cp", "UtoC", cp_site_proportion[2]))
push!(sites, ("mt", "CtoU", mt_site_proportion[1]))
push!(sites, ("mt", "UtoC", mt_site_proportion[2]))

sp1 = @df sites groupedbar(:organelle, :proportion, group = :direction, bar_position = :stack, xguide = "editing direction")
```

```julia
cp_codon_position = [24.0,188.0,23.0]
cp_codon_position_proportion = cp_codon_position ./= sum(cp_codon_position)

mt_codon_position = [78.0,173.0,27.0]
mt_codon_position_proportion = mt_codon_position ./= sum(mt_codon_position)

codon_positions = DataFrame(organelle = String[], position = String[], proportion = Float64[])
push!(codon_positions, ("cp", "1st", cp_codon_position_proportion[1]))
push!(codon_positions, ("cp", "2nd", cp_codon_position_proportion[2]))
push!(codon_positions, ("cp", "3rd", cp_codon_position_proportion[3]))
push!(codon_positions, ("mt", "1st", mt_codon_position_proportion[1]))
push!(codon_positions, ("mt", "2nd", mt_codon_position_proportion[2]))
push!(codon_positions, ("mt", "3rd", mt_codon_position_proportion[3]))

sp2 = @df codon_positions groupedbar(:organelle, :proportion, group = :position, bar_position = :stack, xguide = "codon position")
```

```julia
cp_syn = [23.0, 212.0]
cp_syn_proportion = cp_syn ./= sum(cp_syn)

mt_syn = [33.0, 245.0]
mt_syn_proportion = mt_syn ./= sum(mt_syn)

synonymous = DataFrame(organelle = String[], cat = String[], proportion = Float64[])
push!(synonymous, ("cp", "synonymous", cp_syn_proportion[1]))
push!(synonymous, ("cp", "non-synonymous", cp_syn_proportion[2]))
push!(synonymous, ("mt", "synonymous", mt_syn_proportion[1]))
push!(synonymous, ("mt", "non-synonymous", mt_syn_proportion[2]))

sp3 = @df synonymous groupedbar(:organelle, :proportion, group = :cat, bar_position = :stack, xguide = "synonmous or not")
```

```julia
cp_minus_one = [38.0,84.0,4.0,159.0]
cp_minus_one_proportion = cp_minus_one ./= sum(cp_minus_one)

mt_minus_one = [43.0,106.0,9.0,204.0]
mt_minus_one_proportion = mt_minus_one ./= sum(mt_minus_one)

minus_ones = DataFrame(organelle = String[], base = String[], proportion = Float64[])
push!(minus_ones, ("cp", "A", cp_minus_one_proportion[1]))
push!(minus_ones, ("cp", "C", cp_minus_one_proportion[2]))
push!(minus_ones, ("cp", "G", cp_minus_one_proportion[3]))
push!(minus_ones, ("cp", "U", cp_minus_one_proportion[4]))
push!(minus_ones, ("mt", "A", mt_minus_one_proportion[1]))
push!(minus_ones, ("mt", "C", mt_minus_one_proportion[2]))
push!(minus_ones, ("mt", "G", mt_minus_one_proportion[3]))
push!(minus_ones, ("mt", "U", mt_minus_one_proportion[4]))


sp4 = @df minus_ones groupedbar(:organelle, :proportion, group = :base, bar_position = :stack, xguide = "base at -1")
```

```julia
plot(sp1, sp2, sp3, sp4, layout = (1, 4), link = :all, size=(1200,300), bottom_margin = 10mm)
```
